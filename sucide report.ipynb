Now create the lineplot to see suicides numbers according to year with age group

```python
age5=df.loc[df.loc[:,'age']=='5-14 years',:]  
age1=df.loc[df.loc[:,'age']=='15-24 years',:]
a=df.loc[df.loc[:,'age']=='25-34 years',:]
ag25=df.loc[df.loc[:,'age']=='35-54 years',:]
ag35=df.loc[df.loc[:,'age']=='55-74 years',:] 
ag7=df.loc[df.loc[:,'age']=='75+ years',:]  


plt.figure(figsize=(16,7))

sns.lineplot(x='year',y='suicides_no',data=age5)
sns.lineplot(x='year',y='suicides_no',data=age1)
sns.lineplot(x='year',y='suicides_no',data=a)
sns.lineplot(x='year',y='suicides_no',data=ag25)
sns.lineplot(x='year',y='suicides_no',data=ag35)
sns.lineplot(x='year',y='suicides_no',data=ag7)

plt.legend(['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years'])
```

```python
male_pop=df.loc[df.loc[:,'sex']=='male',:]
femal_pop=df.loc[df.loc[:,'sex']=='female',:]
sns.lineplot(x='year',y='suicides_no',data=male_pop)
sns.lineplot(x='year',y='suicides_no',data=femal_pop)
plt.legend(['Male','Female'])
```
